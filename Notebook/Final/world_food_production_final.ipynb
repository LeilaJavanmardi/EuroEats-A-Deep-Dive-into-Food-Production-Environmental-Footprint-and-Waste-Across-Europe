```python
#counry, products
grap=fil_WFP.groupby(['Country'])[food].mean().round(2)
grap1=pd.DataFrame(grap).round(2).reset_index()
grap1.describe().round(2)
```

```python
plt.figure(figsize=(20,10))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan', 'xkcd:azure', 'xkcd:lavender', 'xkcd:rose', 'xkcd:peach', 'xkcd:pear', 'xkcd:ocean']

lines_to_show=['Maize Production (tonnes)',
       'Rice  Production ( tonnes)',
       'Wheat Production (tonnes)', 'Tomatoes Production (tonnes)', 'Sweet potatoes  Production (tonnes)',
       'Sugar cane Production (tonnes)',
       'Soybeans  Production (tonnes)',
       'Potatoes  Production (tonnes)',
       'Grapes  Production (tonnes)', 'Meat, chicken  Production (tonnes)']

for i, column in enumerate(grap1.columns):
  if column in lines_to_show:
    plt.plot(grap1['Country'], grap1[column], label=column, color=colors[i % len(colors)])
    plt.xticks(rotation=90)

plt.xlabel('Countries')
plt.ylabel('Production (tonnes)')
plt.title('Production (tonnes) by country')
plt.legend()
```
